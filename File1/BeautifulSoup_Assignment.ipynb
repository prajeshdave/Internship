```python
#Question: 03

area = ["Indiranagar", "Jayanagar", "Rajaji Nagar"]
houses = {}

for locality in area:
    link3 = requests.get("https://www.nobroker.in/")
    soup3 = BeautifulSoup(link3.content)
    list_of_house = []
    
    details = soup3.find_all("div", class_="mx-2.5 nb__2y9P3")
    
    for propty in details:
        title = propty.find("div", class_= "w-100pe flex justify-between items-start relative").text.strip()
        location = propty.find("div", class_="h-1.5p w-1.5p flex flex-shrink-0 justify-center items-center bg-bill-bg-color rounded-5").text.strip()
        price = propty.find("div", class_= "mr-0.2p").text.strip()
        list_of_house.append({
            "Title":title,
            "Location": location,
            "Price": price,            
        })
    houses[locality] = list_of_house
    
for locality, house3 in houses.items():
    print(f"\nHouses in {locality}:\n")
    for house33 in house3:
        print(f"Title: {house33['Title']}")
        print(f"Location: {house33['Location']}")
        print(f"Price: {house33['Price']}")
```

```python
#Question: 04

link4 = requests.get("https://www.bewakoof.com/bestseller?sort=popular")

soup4 = BeautifulSoup(link4.content)

products = soup4.find_all("div", class_="plp-product-card", limit = 10)

ten_products = []

for product in products:
    heading4 = product.find("h3").text.strip()
    price4 = product.find("div", class_="productPriceBox d-flex align-items-end  false")
    img_link = product.find("div", class_= "productImg").text.strip()
    
    ten_products.append({
        "Heading": heading4,
        "Price": price4,
        "Image URL": img_link
    })

for i, product in enumerate(ten_products, 1):
    print(f"Product {i}:")
    print(f"Heading: {product['Heading']}")
    print(f"Price: {product['Price']}")
    print(f"Link: {product['Link']}")
```
