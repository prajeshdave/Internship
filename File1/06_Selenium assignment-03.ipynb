```python
#Answer-04

driver04 = webdriver.Chrome()
driver04.get('https://github.com/')
```

```python
click_explore = driver04.find_element(By.XPATH, '//ul[@class="d-lg-flex list-style-none"]/li[4]/button')
click_explore.click()
click_trending = driver04.find_element(By.XPATH, '//ul[@class="d-lg-flex list-style-none"]/li[4]/div/div[3]/ul/li[2]/a')
click_trending.click()
```

```python
titles = []
descriptions = []
contributors_count = []
languages = []

repository_element = driver04.find_elements(By.XPATH, '//article[@class="Box-row"]')
for repository in repository_element:
    try:
        title_element = repository.find_element(By.XPATH, './/h2[@class="h3 lh-condensed"]') #Using "." at the start of XPATH expression which ensures that elements from current context node only considered and do not consider with subXPATH inside loop
        if title_element is not None:
            titles.append(title_element.text)
    except:
        titles.append("-")

    try:
        description_element = repository.find_element(By.XPATH, './/p[@class="col-9 color-fg-muted my-1 pr-4"]')
        descriptions.append(description_element.text)
    except:
        descriptions.append("-")

    try:
        count_element = repository.find_element(By.XPATH, './/div[@class="f6 color-fg-muted mt-2"]/a[1]')
        if count_element is not None:
            contributors_count.append(count_element.text)
    except:
        contributors_count.append("-")

    try:   
        language_element = repository.find_element(By.XPATH, './/span[@class="d-inline-block ml-0 mr-3"]')
        if language_element is not None:
            languages.append(language_element.text)
    except:
        languages.append("-")
```

```python
print((len(titles), len(descriptions), len(contributors_count), len(languages)))
```

```python
df = pd.DataFrame({'Titles':titles, 'Descriptions': descriptions, "contributor's count":contributors_count, 'Language':languages})
df
```

```python
#Answer-05:

driver05 = webdriver.Chrome()
driver05.get("https://www.billboard.com/")
```

```python
click_chart = driver05.find_element(By.XPATH, '/html/body/div[3]/div[9]/div/div/div/ul/li[1]/ul/li[2]')
click_chart.click()
```

```python
songs_name = []
artists_name = []
last_week_rank = []
peak_rank = []
weeks_onboard = []

songname_element = driver05.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li/h3')
for songname in songname_element:
    songs_name.append(songname.text)
    
artistname_element = driver05.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[1]/span')
for artistname in artistname_element:
    artists_name.append(artistname.text)
    
lastweek_rank_element = driver05.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[4]/span')
for lastweek in lastweek_rank_element:
    last_week_rank.append(lastweek.text)
    
peakrank_element = driver05.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[7]/ul/li[4]')
for peakrank in peakrank_element:
    peak_rank.append(peakrank.text)
    
weeksonboard = driver05.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[7]/ul/li[5]')
for week in weeksonboard:
    weeks_onboard.append(week.text)
```

```python
len(peak_rank)
```

```python
import pandas as pd
df = pd.DataFrame({'Song name':songs_name, 'Artist name':artists_name, 'Last week rank':last_week_rank, 'Peak rank':peak_rank, 'Weeks on board':weeks_onboard})
pd.set_option('display.max_rows',None) #to view all the rows
df
```

```python
#Answer-07:

driver07 = webdriver.Chrome()
driver07.get('https://www.imdb.com/list/ls512407256/')
```

```python
names = []
year_span = []
genres = []
run_timings = []
ratings = []
votes = []

name_element = driver07.find_elements(By.XPATH, '//a[@class="ipc-title-link-wrapper"]/h3')
for name in name_element:
    names.append(name.text)
    
yearspan_element = driver07.find_elements(By.XPATH, '//div[@class="sc-b189961a-7 btCcOY dli-title-metadata"]/span[1]')
for year in yearspan_element:
    year_span.append(year.text)
    
genre_element = driver07.find_elements(By.XPATH, '//div[@class="sc-b189961a-7 btCcOY dli-title-metadata"]/span[3]')
for genre in genre_element:
    genres.append(genre.text)
    
runtime_element = driver07.find_elements(By.XPATH, '//div[@class="sc-b189961a-7 btCcOY dli-title-metadata"]/span[2]')
for runtime in runtime_element:
    run_timings.append(runtime.text)
    
rating_element = driver07.find_elements(By.XPATH, '//span[@class="ipc-rating-star--rating"]')
for rating in rating_element:
    ratings.append(rating.text)
    
vote_element = driver07.find_elements(By.XPATH, '//span[@class="ipc-rating-star--voteCount"]')
for vote in vote_element:
    votes.append(vote.text)
```

```python
df07 = pd.DataFrame({'Name':names, 'Year span':year_span, 'Genre':genres, 'Run time':run_timings, 'Ratings':ratings, 'Votes':votes})
df07
```

```python
#Answer-08

driver08 = webdriver.Chrome()
driver08.get('https://archive.ics.uci.edu/')
```

```python
click_dataset = driver08.find_element(By.XPATH, '//nav[@class="navbar-center hidden lg:flex"]/ul/li[1]/a')
click_dataset.click()
```

```python
click_expand = driver08.find_element(By.XPATH, '//div[@class="swap"][2]')
click_expand.click()
```

```python
dataset_names = []
data_types = []
tasks = []
attribute_types = []
instances = []
attributes = []
years = []

name_element = driver08.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')
for name in name_element:
    dataset_names.append(name.text)
    
datatype_element = driver08.find_elements(By.XPATH, '//div[@class="col-span-3 flex items-center gap-2"][2]')
for datatype in datatype_element:
    data_types.append(datatype.text)
    
task_element = driver08.find_elements(By.XPATH, '//div[@class="col-span-3 flex items-center gap-2"][1]')
for task in task_element:
    tasks.append(task.text)
    
attribute_type_element = driver08.find_elements(By.XPATH, '//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]')
for attribute in attribute_type_element:
    attribute_types.append(attribute.text)
    
instance_element = driver08.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
for instance in instance_element:
    instances.append(instance.text)
    
attribute_no_element = driver08.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
for attribute_nos in attribute_no_element:
    attributes.append(attribute_nos.text)
    
year_element = driver08.find_elements(By.XPATH, '//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[3]')
for year in year_element:
    years.append(year.text)
```

```python
print(len(dataset_names), len(data_types), len(tasks), len(attribute_types), len(instances), len(attributes), len(years))
```

```python
df08 = pd.DataFrame({'Dataset name':dataset_names, 'Data types':data_types, 'Task':tasks, 'Attribute_type':attribute_types, 'No. of Instances':instances, 'No. of Attributes':attributes, 'Year':years})
df08
```
