```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
```

```python
#Answer-01

driver01 = webdriver.Chrome()
driver01.get('https://www.amazon.in/') #opening the given link
```

```python
user_search01 = driver01.find_element(By.XPATH, '//div[@class="nav-search-field "]/input') #search box element
user_search01.send_keys(input()) #taking user input for search value
click01 = driver01.find_element(By.XPATH, '//div[@class="nav-search-submit nav-sprite"]/span/input') #search clicking element
click01.click() #clicking the search button
```

```python
#Answer-02
#considering empty lists for asked parameters
brands = []
products = []
prices = []
returns_exchanges = []
expected_delivery = []
availability = []
product_url = []

for page in range(1,4): #for giving page range of 1 to 3, Note:for selenium when interacting with XPATH, indexing starts with 1 instead of 0
    delay = 5
    url_links = driver01.find_elements(By.XPATH, '//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]/a')
    for link in url_links:
        product_url.append(link.get_attribute('href'))        
    try: #for opening the particular product from the given page
        product_details = WebDriverWait(driver01, delay).until(EC.presence_of_all_elements_located((By.XPATH, '//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]/a'))) #wait for the full loading of page
        for one_product in product_details:
            driver01.execute_script("arguments[0].removeAttribute('target');", one_product) #removing 'target' attribute to avoid opening in new tab
            one_product.click() 
            try: #to scrap the asked parameters from the opened product details
                try:
                    brand_element = driver01.find_elements(By.XPATH, '//tr[@class="a-spacing-small po-brand"]/td[2]/span') #scrapping the  from given page
                    for brand_name in brand_element:
                        brands.append(brand_name.text)
                except NoSuchElementException:
                    brands.append("-") #to add 'None' in the data, if any parameter is missing or not available

                try:
                    product_element = driver01.find_elements(By.XPATH, '//span[@class="a-size-large product-title-word-break"]') #scrapping the  from given page
                    for product in product_element:
                        products.append(product.text)
                except NoSuchElementException:
                    products.append("-")

                try:
                    price_element = driver01.find_elements(By.XPATH, '//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]/span[2]/span[2]') #scrapping the  from given page
                    for price in price_element:
                        prices.append(price.text)
                except NoSuchElementException:
                    prices.append("-")

                try:
                    return_exchange_element = driver01.find_elements(By.XPATH, '//div[@class="a-section icon-farm-margins celwidget"]/div/div/div/div[2]/div/ol/li/div/span/div[2]/span') #scrapping the  from given page
                    for return_exchange in return_exchange_element:
                        returns_exchanges.append(return_exchange.text)
                except NoSuchElementException:
                    returns_exchanges.append("-")

                try:
                    delivery_element = driver01.find_elements(By.XPATH, '//div[@class="a-spacing-base"][1]/span/span[1]') #scrapping the  from given page
                    for approx_delivery_date in delivery_element:
                        expected_delivery.append(approx_delivery_date.text)
                except NoSuchElementException:
                    expected_delivery.append("-")

                try:
                    availability_element = driver01.find_elements(By.XPATH, '//div[@class="a-section a-spacing-none a-spacing-top-micro }"]/span') #scrapping the  from given page
                    for availability_status in availability_element:
                        availability.append(availability_status.text)
                except NoSuchElementException:
                    availability.append("-")
                    
            except NoSuchElementException as e:
                print("Exception Raised: ", e)

            driver01.back() #for coming back to the original page after scraping all the asked parameters from product details
            WebDriverWait(driver01, delay).until(EC.presence_of_element_located((By.XPATH, '//div[@class="sg-col-inner"]/div/span/div/div/div/div[2]/div/div/div/h2/a'))) #wait for the full loading of page
  
    except StaleElementReferenceException as SE: #to handle exception if any mismatch raise for any particular page
        print("Exception raised: ", SE)
        
    time.sleep(5) #waiting period to load fully, after scraping all product details from given page
    next_page_length = driver01.find_elements(By.XPATH, '//span[@class="s-pagination-strip"]/a')
    next_page_condition = len(next_page_length)
    next_button = driver01.find_element(By.XPATH, f'//span[@class="s-pagination-strip"]/a[{next_page_condition}]') #for interacting with next page
    next_button.click()
```

```python
print(len(brands), len(products), len(prices), len(returns_exchanges), len(expected_delivery), len(availability), len(product_url)) #to verify the lenghths of all the parameters
```

```python
import pandas as pd #for creating the datasheet
df = pd.DataFrame({'Brands': brands[0:22], 'Products':products[0:22], 'Prices': prices[0:22], 'Returns/Exchanges':returns_exchanges[0:22], 'Expected Delivery': expected_delivery[0:22], 'Availability': availability[0:22], 'Product url Links':product_url[0:22]})
df
```

```python
df.to_csv('Answer-01-02.csv')
```

```python
#Answer-04:
driver04 = webdriver.Chrome()
driver04.get('https://www.flipkart.com')
```

```python
search04 = driver04.find_element(By.XPATH, '//input[@class="Pke_EE"]')
search04.send_keys('smart phones')
click04 = driver04.find_element(By.XPATH, '//button[@class="_2iLD__"]')
click04.click()
```

```python
brands = []
smartphone_name = []
colour = []
RAM = []
ROM_storage = []
primary_camera = []
secondary_camera = []
display_size = []
battery_capacity = []
prices = []
product_url = []


delay = 5
url_links = driver04.find_elements(By.XPATH, '//a[@class="CGtC98"]')
for link in url_links:
    product_url.append(link.get_attribute('href'))        

try:
    brand_element = driver04.find_elements(By.XPATH, '//div[@class="KzDlHZ"]') #scrapping the  from given page
    for brand_name in brand_element:
        brands.append(brand_name.text)
except NoSuchElementException:
    brands.append("-") #to add 'None' in the data, if any parameter is missing or not available

try:
    product_element = driver04.find_elements(By.XPATH, '//div[@class="KzDlHZ"]') #scrapping the  from given page
    for product in product_element:
        smartphone_name.append(product.text)
except NoSuchElementException:
    smartphone_name.append("-")

try:
    colour_element = driver04.find_elements(By.XPATH, '//div[@class="KzDlHZ"]') #scrapping the  from given page
    for colr in colour_element:
        colour.append(colr.text)
except NoSuchElementException:
    colour.append("-")

try:
    rams = driver04.find_elements(By.XPATH, '//ul[@class="G4BRas"]/li[1]') #scrapping the  from given page
    for ram in rams:
        RAM.append(ram.text)
except NoSuchElementException:
    RAM.append("-")    

try:
    roms = driver04.find_elements(By.XPATH, '//ul[@class="G4BRas"]/li[1]') #scrapping the  from given page
    for rom in roms:
        ROM_storage.append(rom.text)
except NoSuchElementException:
    ROM_storage.append("-")      

try:
    prime_camera = driver04.find_elements(By.XPATH, '//ul[@class="G4BRas"]/li[3]') #scrapping the  from given page
    for p_camera in prime_camera:
        primary_camera.append(p_camera.text)
except NoSuchElementException:
    primary_camera.append("-")    

try:
    second_camera = driver04.find_elements(By.XPATH, '//ul[@class="G4BRas"]/li[3]') #scrapping the  from given page
    for s_camera in second_camera:
        secondary_camera.append(s_camera.text)
except NoSuchElementException:
    secondary_camera.append("-")        

try:
    sizes = driver04.find_elements(By.XPATH, '//ul[@class="G4BRas"]/li[2]') #scrapping the  from given page
    for d_size in sizes:
        display_size.append(d_size.text)
except NoSuchElementException:
    display_size.append("-")       

try:
    power_capacity = driver04.find_elements(By.XPATH, '//ul[@class="G4BRas"]/li[4]') #scrapping the  from given page
    for power in power_capacity:
        battery_capacity.append(power.text)
except NoSuchElementException:
    battery_capacity.append("-")     
    
try:
    price_element = driver04.find_elements(By.XPATH, '//div[@class="Nx9bqj _4b5DiR"]') #scrapping the  from given page
    for price in price_element:
        prices.append(price.text)
except NoSuchElementException:
    prices.append("-")
```

```python
import pandas as pd #for creating the datasheet
df = pd.DataFrame({'Brands': brands, 'Smartphone Name':smartphone_name, 'Colour': colour, 'RAM':RAM, 'ROM Storage': ROM_storage, 'Primary Camera': primary_camera, 'Secondary Camera':secondary_camera, 'Display Size': display_size, 'Battery Capacity': battery_capacity, 'Prices': prices, "Product URL": product_url})
df
```

```python
df.to_csv('Answer-05.csv')
```

```python
#Answer-07:
driver07 = webdriver.Chrome()
driver07.get('https://www.forbes.com/billionaires/')
```

```python
ranks = []
names = []
networths = []
ages = []
citizenships = []
sources = []
industries = []

rankings = driver07.find_elements(By.XPATH, '//div[@class="Table_rank__X4MKf"]/div')
for rank in rankings:
    ranks.append(rank.text)

namings = driver07.find_elements(By.XPATH, '//div[@class="Table_personName__Bus2E"]')
for name in namings:
    names.append(name.text)
    
worths = driver07.find_elements(By.XPATH, '//div[@class="Table_finalWorth__UZA6k"]')
for worth in worths:
    networths.append(worth.text)
    
sourcings = driver07.find_elements(By.XPATH, '//div[@class="Table_finalWorth__UZA6k"]')
for source in sourcings:
    sources.append(source.text)
    
industry_domain = driver07.find_elements(By.XPATH, '//div[@class="Table_finalWorth__UZA6k"]')
for domain in industry_domain:
    industries.append(domain.text)
```

```python
import pandas as pd
df = pd.DataFrame({'Ranks':ranks, 'Names':names, 'Networths':networths, 'Sources':sources, 'Industry':industries})
df
```
